```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json
from urllib.request import urlretrieve
import os

# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r = json.loads(r.text)
print('ä¸€ç­†è³‡æ–™çš„æ¬„ä½æœ‰:')
for i in r[0].keys():
    print(i)
    
print('å…¨éƒ¨å…±æœ‰',len(r), 'ç­†è³‡æ–™')
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for i,v in enumerate(r):
    print('ç¬¬',i+1,'ç­†è³‡æ–™')
    print(v['title'], v['createdAt'], v['commentCount'],'å‰‡ç•™è¨€ï¼Œ', v['likeCount'],'å€‹è®š')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

count = 0
like = 0
for i in r :
    count = count + int(i['commentCount'])
    like = like + int(i['likeCount'])
print('ç†±é–€æ–‡ç« çš„å¹³å‡ç•™è¨€äººæ•¸æ˜¯: ',count /len(r))
print('ç†±é–€æ–‡ç« çš„å¹³å‡æŒ‰è®šäººæ•¸æ˜¯ï¼š',like/ len(r))    
```
